```python
class Student:
    #define class_variable.
    school_name = 'Sakura-highschool'
    total_student = 0
    all_students = []

    def __init__(self, name: str):
        self.name = name
        Student.total_student += 1
        Student.all_students.append(self)
        
    #通常のメソッド(インスタンスに固有の情報)
    def introduce(self):
        print(f'私の名前は{self.name}です')

    #Class method
    @classmethod
    def get_school_inf(cls):
        print(f'学校の名前は{cls.school_name}です')
        print(f'生徒数は{cls.total_student}人です')

    @classmethod
    def change_school_name(cls, new_name: str):
        old_name = cls.school_name
        cls.school_name = new_name
        print(f'学校名を{old_name}から、{new_name}に変更しました')

    
    @classmethod
    def list_all_student(cls):
        print(f'==={cls.school_name}の生徒一覧 ===')
        for student in cls.all_students:
            print(f'- {student.name}')


    
```

```python
print('入学前')
Student.get_school_inf()

print('\n---入学---')
tanaka = Student('Tanaka')
sato = Student('Sato')
yamada = Student('Yamada')

print('\n---学校情報---')
Student.get_school_inf()

print('\n---生徒一覧---')
Student.list_all_student()

print('\n---学校名変更---')
Student.change_school_name('Cherry-Highschool')
print('\n---変更後---')
Student.list_all_student()
```
