```python
class dog:
    #まずは自分の固有の情報(インスタンス変数)を決める。
    #属性アクセスできるように、self.インスタンス変数 = aurgumentで、インスタンス変数を定義
    def __init__(self, name, skill):
        self.namae = name
        self.udemae = skill
    
    #インスタンス化されたオブジェクトの機能であるメソッドを作る
    def aisatsu(self):
        print(f'私の前は{self.namae}です。')
    
    def waza(self):
        print(f'俺は{self.udemae}ができます。')

    def assobu(self, toy):
        print(f'{self.namae}は{toy}で遊ぶのが好きです')

    def oshieru(self,word):
        print(f'{self.namae}は「{word}」と言いました')

colo = dog('colo', 'eat')
ester = dog('ester', 'run')

colo.aisatsu()
colo.waza()
colo.assobu('牛骨')
colo.oshieru('ワンワン')
ester.assobu('牛タン')

```
