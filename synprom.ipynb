```python
"""
人工プロモーター設計ツール
Artificial Promoter Design Tool

このスクリプトは、コアプロモーター配列と転写因子結合部位（TFB）を
柔軟に組み合わせて人工プロモーターを設計するための関数を提供します。
"""

def Coplanar(core_promoter, TFB, spacer_length=10):
    """
    Coplanar型人工プロモーターを設計する関数
    
    TFBをcore_promoterの上流に、指定されたスペーサー長で配置します。
    
    Parameters:
    -----------
    core_promoter : str
        コアプロモーター配列（DNA塩基配列: A, T, G, C）
    TFB : str
        転写因子結合部位の配列（DNA塩基配列: A, T, G, C）
    spacer_length : int, optional
        TFBとcore_promoterの間のスペーサー長（デフォルト: 10塩基）
        10または11塩基を推奨
    
    Returns:
    --------
    str
        設計された人工プロモーター配列（TFB + スペーサー + core_promoter）
    
    Examples:
    ---------
    >>> Coplanar("TATAAA", "GCCAAT", spacer_length=10)
    'GCCAATNNNNNNNNNNNTATAAA'
    """
    # 入力検証
    if not core_promoter or not TFB:
        raise ValueError("core_promoterとTFBは空であってはいけません")
    
    # スペーサー配列を生成（'N'で表現、または特定の配列を指定可能）
    spacer = 'N' * spacer_length
    
    # 人工プロモーター配列を構築: 5' - TFB - spacer - core_promoter - 3'
    artificial_promoter = TFB + spacer + core_promoter
    
    return artificial_promoter


def Opposite(core_promoter, TFB, spacer_length=5):
    """
    Opposite型人工プロモーターを設計する関数
    
    TFBをcore_promoterの上流に、5塩基のスペーサーで配置します。
    
    Parameters:
    -----------
    core_promoter : str
        コアプロモーター配列（DNA塩基配列: A, T, G, C）
    TFB : str
        転写因子結合部位の配列（DNA塩基配列: A, T, G, C）
    spacer_length : int, optional
        TFBとcore_promoterの間のスペーサー長（デフォルト: 5塩基）
    
    Returns:
    --------
    str
        設計された人工プロモーター配列（TFB + スペーサー + core_promoter）
    
    Examples:
    ---------
    >>> Opposite("TATAAA", "GCCAAT")
    'GCCAATNNNNNNTATAAA'
    """
    # 入力検証
    if not core_promoter or not TFB:
        raise ValueError("core_promoterとTFBは空であってはいけません")
    
    # スペーサー配列を生成
    spacer = 'N' * spacer_length
    
    # 人工プロモーター配列を構築: 5' - TFB - spacer - core_promoter - 3'
    artificial_promoter = TFB + spacer + core_promoter
    
    return artificial_promoter


def design_promoter_with_custom_spacer(core_promoter, TFB, spacer_sequence):
    """
    カスタムスペーサー配列を使用して人工プロモーターを設計する関数
    
    Parameters:
    -----------
    core_promoter : str
        コアプロモーター配列
    TFB : str
        転写因子結合部位の配列
    spacer_sequence : str
        特定のスペーサー配列（'N'の代わりに特定の配列を使用）
    
    Returns:
    --------
    str
        設計された人工プロモーター配列
    """
    return TFB + spacer_sequence + core_promoter


def validate_sequence(sequence):
    """
    DNA配列の妥当性を検証する関数
    
    Parameters:
    -----------
    sequence : str
        検証するDNA配列
    
    Returns:
    --------
    bool
        配列が有効な場合True、無効な場合False
    """
    valid_bases = set('ATGCNatgcn')
    return all(base in valid_bases for base in sequence)


# 使用例
if __name__ == "__main__":
    print("=" * 60)
    print("人工プロモーター設計ツール")
    print("=" * 60)
    
    # 例1: Coplanar型プロモーター（10塩基スペーサー）
    print("\n【例1】Coplanar型プロモーター（10塩基スペーサー）")
    core1 = "TATAAATA"
    tfb1 = "GCCAAT"
    promoter1 = Coplanar(core1, tfb1, spacer_length=10)
    print(f"Core Promoter: {core1}")
    print(f"TFB:           {tfb1}")
    print(f"結果:          {promoter1}")
    print(f"長さ:          {len(promoter1)} bp")
    
    # 例2: Coplanar型プロモーター（11塩基スペーサー）
    print("\n【例2】Coplanar型プロモーター（11塩基スペーサー）")
    promoter2 = Coplanar(core1, tfb1, spacer_length=11)
    print(f"結果:          {promoter2}")
    print(f"長さ:          {len(promoter2)} bp")
    
    # 例3: Opposite型プロモーター（5塩基スペーサー）
    print("\n【例3】Opposite型プロモーター（5塩基スペーサー）")
    core2 = "TATAAA"
    tfb2 = "CAAT"
    promoter3 = Opposite(core2, tfb2)
    print(f"Core Promoter: {core2}")
    print(f"TFB:           {tfb2}")
    print(f"結果:          {promoter3}")
    print(f"長さ:          {len(promoter3)} bp")
    
    # 例4: カスタムスペーサーを使用
    print("\n【例4】カスタムスペーサー配列を使用")
    custom_spacer = "ATATATAT"
    promoter4 = design_promoter_with_custom_spacer(core2, tfb2, custom_spacer)
    print(f"Custom Spacer: {custom_spacer}")
    print(f"結果:          {promoter4}")
    print(f"長さ:          {len(promoter4)} bp")
    
    # 複数のプロモーター設計例
    print("\n【例5】複数のプロモーター設計")
    core_variants = ["TATAAA", "TATAAT", "TTGACA"]
    tfb_variants = ["CAAT", "GCCAAT", "CCAAT"]
    
    for i, (core, tfb) in enumerate(zip(core_variants, tfb_variants), 1):
        coplanar = Coplanar(core, tfb, spacer_length=10)
        opposite = Opposite(core, tfb)
        print(f"\nバリアント {i}:")
        print(f"  Coplanar: {coplanar}")
        print(f"  Opposite: {opposite}")
    
    print("\n" + "=" * 60)
```

```python
# アノテーション削除版 - プロモーターライブラリ生成対応

import pandas as pd
from itertools import product

def Coplanar(core_promoter, TFB, spacer_length=10):
    if not core_promoter or not TFB:
        raise ValueError("core_promoterとTFBは空であってはいけません")
    
    spacer = 'N' * spacer_length
    artificial_promoter = TFB + spacer + core_promoter
    
    return artificial_promoter


def Opposite(core_promoter, TFB, spacer_length=5):
    if not core_promoter or not TFB:
        raise ValueError("core_promoterとTFBは空であってはいけません")
    
    spacer = 'N' * spacer_length
    artificial_promoter = TFB + spacer + core_promoter
    
    return artificial_promoter


def design_promoter_with_custom_spacer(core_promoter, TFB, spacer_sequence):
    return TFB + spacer_sequence + core_promoter


def validate_sequence(sequence):
    valid_bases = set('ATGCNatgcn')
    return all(base in valid_bases for base in sequence)


def generate_promoter_library(core_csv=None, tfb_csv=None, 
                               core_list=None, tfb_list=None,
                               promoter_types=['Coplanar', 'Opposite'],
                               coplanar_spacer=10, opposite_spacer=5,
                               output_csv=None):
    """
    プロモーターライブラリを生成する関数
    
    CSVファイルまたはリストからcore_promoterとTFBを読み込み、
    全ての組み合わせで人工プロモーターを生成します。
    """
    
    # core_promoterの読み込み
    if core_csv:
        core_df = pd.read_csv(core_csv)
        # 'sequence'カラムがあると仮定、なければ最初のカラムを使用
        core_col = 'sequence' if 'sequence' in core_df.columns else core_df.columns[0]
        cores = core_df[core_col].tolist()
    elif core_list:
        cores = core_list
    else:
        raise ValueError("core_csvまたはcore_listを指定してください")
    
    # TFBの読み込み
    if tfb_csv:
        tfb_df = pd.read_csv(tfb_csv)
        tfb_col = 'sequence' if 'sequence' in tfb_df.columns else tfb_df.columns[0]
        tfbs = tfb_df[tfb_col].tolist()
    elif tfb_list:
        tfbs = tfb_list
    else:
        raise ValueError("tfb_csvまたはtfb_listを指定してください")
    
    # 全ての組み合わせを生成
    library = []
    
    for i, (core, tfb) in enumerate(product(cores, tfbs), 1):
        entry = {
            'ID': f'Promoter_{i:05d}',
            'Core_Promoter': core,
            'TFB': tfb,
        }
        
        if 'Coplanar' in promoter_types:
            coplanar_seq = Coplanar(core, tfb, spacer_length=coplanar_spacer)
            entry['Coplanar_Sequence'] = coplanar_seq
            entry['Coplanar_Length'] = len(coplanar_seq)
        
        if 'Opposite' in promoter_types:
            opposite_seq = Opposite(core, tfb, spacer_length=opposite_spacer)
            entry['Opposite_Sequence'] = opposite_seq
            entry['Opposite_Length'] = len(opposite_seq)
        
        library.append(entry)
    
    # DataFrameに変換
    library_df = pd.DataFrame(library)
    
    # CSVに保存（オプション）
    if output_csv:
        library_df.to_csv(output_csv, index=False)
        print(f"ライブラリを {output_csv} に保存しました")
    
    return library_df


# 使用例
if __name__ == "__main__":
    print("=" * 60)
    print("人工プロモーターライブラリ生成ツール")
    print("=" * 60)
    
    # サンプルデータでテスト
    print("\n【サンプルデータでライブラリ生成】")
    
    sample_cores = ["TATAAA", "TATAAT", "TTGACA", "TATAAATA"]
    sample_tfbs = ["CAAT", "GCCAAT", "CCAAT"]
    
    library = generate_promoter_library(
        core_list=sample_cores,
        tfb_list=sample_tfbs,
        promoter_types=['Coplanar', 'Opposite'],
        coplanar_spacer=10,
        opposite_spacer=5,
        output_csv='promoter_library.csv'  # CSVファイルに保存
    )
    
    print(f"\n生成されたプロモーター数: {len(library)}")
    print(f"組み合わせ: {len(sample_cores)} cores × {len(sample_tfbs)} TFBs = {len(library)} promoters")
    print("\n最初の5件:")
    print(library.head())
    
    print("\n統計情報:")
    if 'Coplanar_Length' in library.columns:
        print(f"Coplanar型の長さ範囲: {library['Coplanar_Length'].min()} - {library['Coplanar_Length'].max()} bp")
    if 'Opposite_Length' in library.columns:
        print(f"Opposite型の長さ範囲: {library['Opposite_Length'].min()} - {library['Opposite_Length'].max()} bp")
    
    print("\n" + "=" * 60)
    print("\nCSVファイルからの読み込み例:")
    print("# core_promoters.csv と tfb_sites.csv を用意して:")
    print("library = generate_promoter_library(")
    print("    core_csv='core_promoters.csv',")
    print("    tfb_csv='tfb_sites.csv',")
    print("    output_csv='my_promoter_library.csv'")
    print(")")
    print("=" * 60)
```
